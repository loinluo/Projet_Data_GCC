```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.fft import fft, fftfreq
import os
import time
```

```python
#fonctions et variables utiles

def read_text_file(file_path):
    """
    Entrée:
        file_path: fichier .txt ou .blochet lu ligne par ligne
    Sortie:
        lines: list des lignes 
    """
    with open(file_path, encoding='ISO-8859-1') as f:
        lines = f.readlines()
        return lines
```

```python
def lines_to_list(lines, list_data):
    """
    Entrée:
        lines: list des lignes
        list_data: list of lists storing the datas
    Sortie:
        list_data: list of lists added the datas in the lines 
    """
    data = list_data
    n = len(lines) // 629
    for i in range(n):
        sub_lines = lines[i * 629: i * 629 + 629]
        list_t = []
        list_v = []
        list_f = []
        list_ff = []
        list_fi = []
        for j in range(15, 629):
            list_t.append(float(sub_lines[j].split()[0]))
            list_v.append(float(sub_lines[j].split()[1]))
            list_f.append(float(sub_lines[j].split()[2]))
            list_ff.append(float(sub_lines[j].split()[3]))
            list_fi.append(float(sub_lines[j].split()[4]))
        
        T = 1./10000 # 10000
        xf = fftfreq(len(list_v), T)
        yf = np.abs(fft(list_v))
        xf = xf.tolist()
        yf = yf.tolist()

        data[0].append(float(sub_lines[6].split()[1]))
        data[1].append(float(sub_lines[9].split()[1]))
        data[2].append(float(sub_lines[10].split()[1]))
        data[3].append(float(sub_lines[11].split()[1]))
        data[4].append(float(sub_lines[12].split()[1]))
        data[5].append(list_t)
        data[6].append(list_v)
        data[7].append(list_f)
        data[8].append(list_ff)
        data[9].append(list_fi)
        if i == n - 1 and len(lines[-1].split()) >= 2:
            data[10].append(float(lines[-1].split()[1]))
        else:
            data[10].append(None)
        data[11].append(xf[:len(list_v)//2])
        data[12].append(yf[:len(list_v)//2])
        
    return data
```
